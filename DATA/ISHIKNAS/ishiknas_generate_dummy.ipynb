```python
def random_date(start, end):
    delta = end - start
    random_days = random.randint(0, delta.days)
    return start + timedelta(days=random_days)

start_date = datetime(2024, 1, 1)
end_date = datetime(2025, 12, 31)
```

```python
data_mutasi = []

for i in range(500):
    row = {
        "id_mutasi": f"MUT-{i+1:04d}",
        "tgl_mutasi": random_date(start_date, end_date).strftime("%m/%d/%y"),
        "id_asal": random.choice(ref_wilayah["id_wilayah"]),
        "id_tujuan": random.choice(ref_wilayah["id_wilayah"]),
        "id_hewan": random.choice(ref_hewan["id_hewan"]),
        "jumlah_ekor": random.randint(10, 200)
    }

    # ❗ Inject missing value (5%)
    if random.random() < 0.05:
        row["jumlah_ekor"] = None

    data_mutasi.append(row)

tr_mutasi = pd.DataFrame(data_mutasi)
tr_mutasi.head()
```

```python
data_sakit = []

for i in range(300):
    row = {
        "id_laporan": f"LAP-{i+1:04d}",
        "tgl_lapor": random_date(start_date, end_date).strftime("%m/%d/%y"),
        "id_wilayah": random.choice(ref_wilayah["id_wilayah"]),
        "id_hewan": random.choice(ref_hewan["id_hewan"]),
        "jumlah_gejala": random.randint(1, 50)
    }

    # ❗ Missing value (10%)
    if random.random() < 0.1:
        row["jumlah_gejala"] = None

    data_sakit.append(row)

tr_laporan_sakit = pd.DataFrame(data_sakit)
tr_laporan_sakit.head()
```

```python
penyakit_map = {
    "H4-001": ["PMK", "LSD", "Anthrax"],
    "H2-001": ["Flu Burung", "Newcastle Disease"]
}

data_lab = []

for i in range(300):
    laporan = random.choice(tr_laporan_sakit["id_laporan"].tolist())
    hewan = random.choice(ref_hewan["id_hewan"])

    row = {
        "id_uji": f"LAB-{i+1:04d}",
        "id_laporan": laporan,
        "tgl_uji": random_date(start_date, end_date).strftime("%m/%d/%y"),
        "penyakit": random.choice(penyakit_map[hewan]),
        "hasil_uji": random.choice(["Positif", "Negatif"])
    }

    # ❗ Missing value di hasil_uji (7%)
    if random.random() < 0.07:
        row["hasil_uji"] = None

    data_lab.append(row)

tr_hasil_lab = pd.DataFrame(data_lab)
tr_hasil_lab.head()
```

```python
data_rph = []

for i in range(400):
    hewan = random.choice(ref_hewan["id_hewan"])

    row = {
        "id_potong": f"PTG-{i+1:04d}",
        "tgl_potong": random_date(start_date, end_date).strftime("%m/%d/%y"),
        "id_wilayah": random.choice(ref_wilayah["id_wilayah"]),
        "id_hewan": hewan,
        "berat_karkas": round(random.uniform(5, 300), 2)
    }

    # ❗ Missing value (5%)
    if random.random() < 0.05:
        row["berat_karkas"] = None

    data_rph.append(row)

tr_rph = pd.DataFrame(data_rph)
tr_rph.head()
```

```python
ref_hewan.to_csv("ref_hewan.csv", index=False)
ref_wilayah.to_csv("ref_wilayah.csv", index=False)
tr_mutasi.to_csv("tr_mutasi.csv", index=False)
tr_laporan_sakit.to_csv("tr_laporan_sakit.csv", index=False)
tr_hasil_lab.to_csv("tr_hasil_lab.csv", index=False)
tr_rph.to_csv("tr_rph.csv", index=False)

print("Semua data berhasil disimpan!")
```

```python
import sqlite3
import pandas as pd

def generate_sql_script(filename, tables_dict):
    with open(filename, 'w') as f:
        f.write("SET NAMES utf8mb4;\n")
        f.write("SET FOREIGN_KEY_CHECKS = 0;\n\n")

        for table_name, df in tables_dict.items():
            f.write(f'DROP TABLE IF EXISTS `{table_name}`;\n')

            # Create table syntax with backticks and MySQL friendly types
            schema = pd.io.sql.get_schema(df, table_name)
            schema = schema.replace('"', '`') # Replace double quotes with backticks
            schema = schema.replace('INTEGER', 'INT')
            schema = schema.replace('REAL', 'DOUBLE')
            f.write(f'{schema};\n\n')

            # Insert statements
            for _, row in df.iterrows():
                cols = '`, `'.join(row.index)
                vals = []
                for val in row.values:
                    if pd.isna(val):
                        vals.append('NULL')
                    elif isinstance(val, (int, float)):
                        vals.append(str(val))
                    else:
                        # Escape single quotes for SQL string
                        clean_val = str(val).replace("'", "''")
                        vals.append(f"'{clean_val}'")

                vals_str = ', '.join(vals)
                f.write(f"INSERT INTO `{table_name}` (`{cols}`) VALUES ({vals_str});\n")
            f.write("\n")

        f.write("SET FOREIGN_KEY_CHECKS = 1;\n")

tables = {
    'ref_hewan': ref_hewan,
    'ref_wilayah': ref_wilayah,
    'tr_mutasi': tr_mutasi,
    'tr_laporan_sakit': tr_laporan_sakit,
    'tr_hasil_lab': tr_hasil_lab,
    'tr_rph': tr_rph
}

generate_sql_script('data_peternakan.sql', tables)
print("File 'data_peternakan.sql' telah diperbarui dengan format MySQL! Silakan download kembali.")
```
